```python
#dependencies and setup
import requests 
from bs4 import BeautifulSoup 
import pandas as pd
# from pprint import pprint
# import os
# import re
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
import time
```

```python
# create a browser instance using splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
time.sleep(1)
```

```python
names = []
mngr_tnrs = []
incptn_dts = []
assets = []
trnovr_rts = []
minimum_investments = []
loads = []
net_expense_ratios = []
gross_expense_ratios = []

for i in range(1,98):
    fidelity_url = f"https://fundresearch.fidelity.com/fund-screener/results/table/management-fees/averageAnnualReturnsYear3/desc/{i}?assetClass=&category=&order=assetClass%2Ccategory"
    browser.visit(fidelity_url)
    time.sleep(4)

    # create HTML object
    html = browser.html

    # parse HTML with BeautifulSoup
    soup = BeautifulSoup(html, 'html.parser')

    div = soup.find('div', id ='static-table-container')
    table = div.find('table', id = 'static-table')
    tbody = table.find('tbody', id = 'static-tbody')
    for listing in tbody.find_all('td', class_ = 'name left'):
        for name in listing.find_all('a'):
            names.append(name.text)

    div2 = soup.find('div', id ='scrollable-results-table-wrapper')
    table2 = div2.find('table', id = 'scrollable-results-table')
    tbody2 = table2.find('tbody', id = 'results-tbody')
    for listing in tbody2.find_all('td', class_ = "managerTenure right"):
        if type(listing.find('span')) == type(None):
            mngr_tnrs.append("")
        else:
            mngr_tnrs.append(listing.find('span').text)
    for listing in tbody2.find_all('td', class_ = "inceptionDate right"):
        if type(listing.find('span')) == type(None):
            assets.append("")
        else:
            assets.append(listing.find('span').text)
    for listing in tbody2.find_all('td', class_ = "assets right"):
        if type(listing.find('span')) == type(None):
            incptn_dts.append("")
        else:
            incptn_dts.append(listing.find('span').text)
    for listing in tbody2.find_all('td', class_ = "turnoverRates right"):
        if type(listing.find('span')) == type(None):
            trnovr_rts.append("")
        else:
            trnovr_rts.append(listing.find('span').text)
    for listing in tbody2.find_all('td', class_ = "minimumInvestment right"):
        if type(listing.find('span')) == type(None):
            minimum_investments.append("")
        else:
            minimum_investments.append(listing.find('span').text)
    for listing in tbody2.find_all('td', class_ = "load right"):
        if type(listing.find('span')) == type(None):
            loads.append("")
        else:
            loads.append(listing.find('span').text)
    for listing in tbody2.find_all('td', class_ = "netExpenseRatio right"):
        if type(listing.find('span')) == type(None):
            net_expense_ratios.append("")
        else:
            net_expense_ratios.append(listing.find('span').text)
    for listing in tbody2.find_all('td', class_ = "grossExpenseRatio right"):
        if type(listing.find('span')) == type(None):
            gross_expense_ratios.append("")
        else:
            gross_expense_ratios.append(listing.find('span').text)
    print(f"{len(names)} funds scraped until page {i}")
```

```python
# close the browser session    
browser.quit()
```

```python
df_mgt_fee = pd.DataFrame({'name' : names, 'manager_tenure': mngr_tnr, 'inception_date': incptn_dts, 'asset_in_mil': assets,\
                            'turnover_rates': trnovr_rts, 'minimum_investment': minimum_investments, 'load': loads,\
                       'net_expense_ratio':net_expense_ratios, 'gross_expense_ratio': gross_expense_ratios})
df_mgt_fee.head()
```
