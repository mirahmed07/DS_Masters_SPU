```python
# get all the csv filepaths and create a list of files
bikedata_2020_02 = "JC-202002-citibike-tripdata.csv"
bikedata_2020_03 = "JC-202003-citibike-tripdata.csv"
bikedata_2020_04 = "JC-202004-citibike-tripdata.csv"
bikedata_2020_11 = "JC-202011-citibike-tripdata.csv"
bikedata_2020_12 = "JC-202012-citibike-tripdata.csv"


all_files = [bikedata_2020_02, bikedata_2020_03, bikedata_2020_04, bikedata_2020_11, bikedata_2020_12]
```

```python
# create empty list to add dataframes to
bikedata_list = []

# for loop to go through each file and read into dataframe, then append to the new list
for file in all_files:
    df = pd.read_csv(file)
    bikedata_list.append(pd.read_csv(file))

# concatenate the list of dataframes to a single dataframe
bikedata_2020 = pd.concat(bikedata_list, sort=False)
bikedata_2020
```

```python
# export the dataframe to a csv file for using in tableau

bikedata_2020.to_csv("all_bikedata_2020.csv", index=False)
```
