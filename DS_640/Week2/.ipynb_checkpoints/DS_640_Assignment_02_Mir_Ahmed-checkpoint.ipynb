```python
# import packages
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
```

# Reading the CSV

```python
# import the csv as a dataframe
df = pd.read_csv('USA_Housing.csv')
df.head()
```

```python
# data type of all columns
df.info()
```

```python
# summary of all numerical variables
df.describe()
```

```python
# Create the correlation heatmap using seaborn
corr = df.corr()
plt.figure(figsize=(25, 25))
sns.set(font_scale=1.5)
sns.heatmap(corr, 
            cmap='viridis', vmax=1.0, vmin=-1.0, linewidths=0.1,
            annot=True, annot_kws={"size": 18}, square=True);
plt.yticks(rotation=0)
plt.tight_layout()
```

# Linear Regression Model Fitting

```python
# Feature selection and train-test splitting
X = df.drop(['Price', 'Address'], axis=1)
y = df["Price"]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# Create a StandardScaler model for both X and y features
X_scaler = StandardScaler().fit(X_train)

# As y variable needs to be a 2D array
y_scaler = StandardScaler().fit(y_train.values.reshape(-1,1))

# Transform using the StandardScaler model
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)

# As y variable needs to be a 2D array
y_train_scaled = y_scaler.transform(y_train.values.reshape(-1,1))
y_test_scaled = y_scaler.transform(y_test.values.reshape(-1,1))

# Shape of features
print(X_train_scaled.shape, y_train_scaled.shape)

# fit scaled feature to the training data
model = LinearRegression()
model.fit(X_train_scaled, y_train_scaled)
```

# Intercept and Coefficients of the Model

```python
# Intercept
print(f'Intercept = {model.intercept_[0]}')
# Coefficients
print(f'Coefficients = {model.coef_[0]}')
```

```python
# scatter plot of residuals of training and testing
plt.scatter(model.predict(X_train_scaled), model.predict(X_train_scaled) - y_train_scaled,\
            c="blue", label="Training Data")
plt.scatter(model.predict(X_test_scaled), model.predict(X_test_scaled) - y_test_scaled, \
            c="orange", label="Testing Data")
plt.legend()
plt.hlines(y=0, xmin=y_test_scaled.min(), xmax=y_test_scaled.max())
plt.title("Residual Plot")
plt.show()
```

# Residual Distribution Histogram

```python
fig1,ax1 = plt.subplots(nrows = 2, figsize = (16,10))
sns.histplot(model.predict(X_train_scaled) - y_train_scaled, ax = ax1[0], \
             kde = True).set(title='Training Residuals')
sns.histplot(model.predict(X_test_scaled) - y_test_scaled, ax = ax1[1], \
             kde = True).set(title='Testing Residuals')
```

# Model Metrics

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

# predictions on test data
predictions = model.predict(X_test_scaled)

# Mean Squared Error
MSE = mean_squared_error(y_test_scaled, predictions)

# Mean Absolute Error
MAE = mean_absolute_error(y_test_scaled, predictions)

# Root Mean Squared Error
RMSE = mean_squared_error(y_test_scaled, predictions, squared=False)

# R-Squared Value
r2 = model.score(X_test_scaled, y_test_scaled)

print(f"MSE:  {MSE}\nMAE:  {MAE}\nRMSE: {RMSE}\nR2:   {r2}")
```
