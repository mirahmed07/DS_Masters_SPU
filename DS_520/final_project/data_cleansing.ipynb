```python
import pandas as pd
import re
import datetime as dt
```

# Overview

```python
df_overview= pd.read_csv('fidelity_mutual_fund_overview.csv')
df_overview.head()
```

```python
df_overview.info()
```

```python
for col in df_overview.columns[2:-1]:
    df_overview[f"{col}"] = df_overview[f"{col}"].str.replace('%', '', regex=True)
    df_overview[f"{col}"] = df_overview[f"{col}"].str.replace('+', '', regex=True)
    df_overview[f"{col}"] = pd.to_numeric(df_overview[f"{col}"], errors='coerce')
```

```python
df_overview.info()
```

```python
df_overview.head()
```

```python
df_overview.describe()
```

# Risk

```python
df_risk.info()
```

# Daily yield

```python
df_yield= pd.read_csv('fidelity_mutual_fund_daily.csv')
df_yield.head()
```

```python
df_yield.columns
```

```python
df_yield.drop(columns=['morningstar_category','nav_dollar_amount', 'nav_change_dollar_amount', 'daily_30_day_yield', 'daily_7_day_yield', 'morningstar_rating_overall'], inplace = True)
```

```python
df_yield.info()
```

```python
for col in df_yield.columns[1:]:
    df_yield[f"{col}"] = df_yield[f"{col}"].str.replace('$', '', regex=True)
    df_yield[f"{col}"] = df_yield[f"{col}"].str.replace(',', '', regex=True)
    df_yield[f"{col}"] = pd.to_numeric(df_yield[f"{col}"], errors='coerce')
```

```python
df_yield.info()
```

```python
df_yield.head()
```

# Merging the dataframes

```python
df = pd.merge(df_overview, df_risk, on = "name", how = "left")
df.head()
```

```python
del df['morningstar_category_y']
```

```python
df.rename(columns={"morningstar_category_x": "morningstar_category"}, inplace = True)
```

```python
df = pd.merge(df, df_yield, on = "name", how = "left")
```

```python
df.head()
```

```python
df.info()
```
