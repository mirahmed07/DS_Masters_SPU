# Airline Dataset Variable definition

| Serial No. | Name | Description |

|18| Dest| destination IATA airport code|
|19| Distance| Distance between airports (miles)|
|20| TaxiIn| Wheels down and arrival at the destination airport gate, in minutes|

# Data Cleansing

```
# checking data types of each column
df.dtypes
```

```
# Changing the data types of DIstance, ArrDelay, DepDelay to integer
df = df.withColumn("Distance", df["Distance"].cast(IntegerType())).withColumn("ArrDelay", df["ArrDelay"].cast(IntegerType())).\
                withColumn("DepDelay", df["DepDelay"].cast(IntegerType()))
df.dtypes
```

```
# five number summary of all numeric columns in the dataframe
numeric_features = [t[0] for t in df.dtypes if t[1] == 'int']
df.select(numeric_features).describe().toPandas().transpose()
```

#### *Columns - CRSElapsedTime and ArrDelay has some null values. We will replace those with 0.*

```
# replacing nulls with 0
df = df.fillna(value=0)
```

```
# Checking for null values in every column
from pyspark.sql.functions import isnan, when, count, col

df.select([count(when(isnan(c), c)).alias(c) for c in df.columns]).show()
```

**Now let's find a specific airlines which has the most delays to flight ratio.**

```
# Aggregation of distance and delays grouped by UniqueCarrier
display(df.groupBy('UniqueCarrier').agg(F.count('Year').alias("count"), F.sum('Distance').alias('Distance'), F.sum('ArrDelay').alias('ArrDelay')\
                                          , F.sum('DepDelay').alias('DepDelay')).\
        withColumn("DelayRatio", (F.col("DepDelay") + F.col("ArrDelay"))/ F.col("count")).orderBy(F.desc('DelayRatio'),F.desc('count')))
```

**AA is the code for American Airlines. So, they had the most delays to flights ratio.**

```
# filtering on American Airlines records only
df = df.filter(df['UniqueCarrier'] == 'AA')
# df.show(n=3)
```

```
# getting the number of partitions for the dataframe
df.rdd.getNumPartitions()
```

```
# Creating a total delay column by adding arrival delay and departure delay
df = df.withColumn('TotalDelay', F.col("ArrDelay") + F.col("DepDelay"))
# df.show(n=3)
```

# Modelling Pre-processing

```
# Stratified Sampling by IsDepDelayed column
df_sampl_strat = df.sampleBy("IsDepDelayed", fractions={'YES':.01, 'NO':.01},seed=42)
df_sampl_strat.dtypes
```

```
# getting counts
df_sampl_strat.count()
```

```
# Removing unnecessary columns
df_sampl_dummy = df_sampl_strat[['Month', 'DayofMonth', 'DayOfWeek', 'CRSDepTime', 'CRSArrTime', 'CRSElapsedTime', 'Origin', 'Dest', 'Distance', \
                                 'DepDelay', 'IsDepDelayed']]
df_sampl_dummy.dtypes
```

```
# check 2 rows
display(df_sampl_dummy.limit(2))
```

```
# convert departure and arrival time columns to hourly categoricaL columns
df_sampl_dummy = df_sampl_dummy.withColumn('CRSDepTime', F.floor(F.col('CRSDepTime')/100).cast(IntegerType())).\
                    withColumn('CRSArrTime', F.floor(F.col('CRSArrTime')/100).cast(IntegerType()))
```

```
# check 2 rows
display(df_sampl_dummy.limit(2))
```

```
# convert to pandas dataframe
df_sampl_dummy = df_sampl_dummy.toPandas()
df_sampl_dummy.info()
```

```
# get dummies for all categorical columns
dummy_df = pd.get_dummies(df_sampl_dummy, columns = ['Month', 'DayofMonth', 'DayOfWeek', 'Origin', 'Dest','CRSDepTime', 'CRSArrTime'])
dummy_df.head().append(dummy_df.tail())
```

```
# convert boolean to numeric column
dummy_df['label'] = dummy_df['IsDepDelayed'].map({'YES': 1, 'NO': 0})
dummy_df.head().append(dummy_df.tail())
```

```
# converting pandas uint8 to int32 for spark's convenience
import numpy as np
dummy_df[dummy_df.select_dtypes('uint8').columns] = dummy_df[dummy_df.select_dtypes('uint8').columns].astype(np.int32)
dummy_df.dtypes
```

```
# shape of the pivoted dataset
dummy_df.shape
```

```
# convert dummy/ pivoted dataframe to spark dataframe
model_df = spark.createDataFrame(dummy_df)
```

```
# creating feature vector
from pyspark.ml.feature import VectorAssembler
assembler = VectorAssembler(inputCols=[x for x in model_df.columns if x not in ['IsDepDelayed', 'DepDelay', 'label']],\
                            outputCol="features",handleInvalid="skip")
# Create a and run a data processing Pipeline
from pyspark.ml import Pipeline
data_prep_pipeline= Pipeline(stages=[assembler])
# Fit and transform the pipeline
cleaner = data_prep_pipeline.fit(model_df)
cleaner_Data = cleaner.transform(model_df)
```

## alternate method

```
# creating a new column with concatenation of origin and destination airport codes
model_df = df_sampl_strat.withColumn('text', F.concat_ws('-', df.Origin, df.Dest))
model_df.dtypes
```

```
# dropping unnecessary columns
model_df = model_df[['Month', 'DayofMonth', 'DayOfWeek', 'CRSDepTime', 'CRSArrTime', 'CRSElapsedTime', 'Distance', 'text', 'IsDepDelayed', 'ArrDelay', 'DepDelay']]
```

```
# Create all the features to the data set
from pyspark.ml.feature import OneHotEncoder, StringIndexer
yes_no_to_num = StringIndexer(inputCol='IsDepDelayed', outputCol='label')
stringIndexer = StringIndexer(inputCol = 'text', outputCol = 'textIndex')
encoder = OneHotEncoder(inputCol='textIndex', outputCol="textEncoded")
# Create feature vectors
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.linalg import Vector
assembler = VectorAssembler(inputCols=['Month', 'DayofMonth', 'DayOfWeek', 'CRSDepTime', 'CRSArrTime', 'CRSElapsedTime', 'Distance', 'textEncoded'],\
                            outputCol="features",handleInvalid="skip")
# Create a and run a data processing Pipeline
from pyspark.ml import Pipeline
data_prep_pipeline= Pipeline(stages=[yes_no_to_num, stringIndexer, encoder, assembler])
```

```
# Fit and transform the pipeline
cleaner = data_prep_pipeline.fit(model_df)
cleaner_Data = cleaner.transform(model_df)
```

# Train Test Split

```
# Break data down into a training set and a testing set
# training, testing = cleaner_Data.select(['Month', 'DayofMonth', 'DayOfWeek', 'CRSDepTime', 'CRSArrTime', 'CRSElapsedTime',\
#                                           'Distance', 'text', 'label', 'features','DepDelay']).randomSplit([0.7,0.3],42)

training, testing = cleaner_Data.randomSplit([0.7,0.3],42)
# saving both training and test on cache
training.cache()
testing.cache()
```

```
# print row counts for training and testing dataset
print(f'Training count: {training.count()}')
print(f'Testing count: {testing.count()}')
```

```
display(training.limit(2))
```

# Naive-Bayes Model

```
from pyspark.ml.classification import NaiveBayes, NaiveBayesModel
from pyspark.ml.evaluation import BinaryClassificationEvaluator
```

```
# Create a Naive Bayes model and fit training data
nb = NaiveBayes()
nb_model = nb.fit(training)
# Tranform the model with the testing data
nb_predictions = nb_model.transform(testing)
```

```
# getting accuracy of the model
evaluator = BinaryClassificationEvaluator(rawPredictionCol="prediction")
accuracy = evaluator.evaluate(nb_predictions)
print ("Model Accuracy: ", accuracy)
```

# Logistic Regression Model

```
from pyspark.ml.classification import LogisticRegression
logReg = LogisticRegression(featuresCol = 'features', labelCol = 'label', maxIter=100)
logRegModel = logReg.fit(training)
```

```
# ROC Curve and AUC for the model
trainingSummary = logRegModel.summary
roc = trainingSummary.roc.toPandas()
plt.plot(roc['FPR'],roc['TPR'])
plt.ylabel('False Positive Rate')
plt.xlabel('True Positive Rate')
plt.title('ROC Curve')
plt.show()
print('Training set areaUnderROC: ' + str(trainingSummary.areaUnderROC))
```

```
# getting accuracy of the model
logReg_predictions = logRegModel.transform(testing)
evaluator = BinaryClassificationEvaluator()
print('Test Area Under ROC', evaluator.evaluate(logReg_predictions))
```

# Decision Tree Classifier

```
from pyspark.ml.classification import DecisionTreeClassifier
dt = DecisionTreeClassifier(featuresCol = 'features', labelCol = 'label', maxDepth = 8, seed = 956191873026065186)
dtModel = dt.fit(training)
dt_predictions = dtModel.transform(testing)
evaluator = BinaryClassificationEvaluator()
print("Test Area Under ROC: " + str(evaluator.evaluate(dt_predictions, {evaluator.metricName: "areaUnderROC"})))
```

# Random Forest Classifier

```
from pyspark.ml.classification import RandomForestClassifier

rf = RandomForestClassifier(featuresCol = 'features', labelCol = 'label', maxDepth = 30, numTrees = 80, seed = 5387697053847413545)
rfModel = rf.fit(training)
rf_predictions = rfModel.transform(testing)
evaluator = BinaryClassificationEvaluator()
print("Test Area Under ROC: " + str(evaluator.evaluate(rf_predictions, {evaluator.metricName: "areaUnderROC"})))
```

# Gradient-Boosted Tree Classifier

```
from pyspark.ml.classification import GBTClassifier
gbt = GBTClassifier(maxIter=50, maxDepth = 15, maxBins = 32, seed = 3504127614838123891)
gbtModel = gbt.fit(training)
gb_predictions = gbtModel.transform(testing)
evaluator = BinaryClassificationEvaluator()
print("Test Area Under ROC: " + str(evaluator.evaluate(gb_predictions, {evaluator.metricName: "areaUnderROC"})))
```

# Linear Regression Model

```
from pyspark.ml.regression import LinearRegression
lr = LinearRegression(featuresCol='features', labelCol='DepDelay')
lrModel = lr.fit(training)
print("Coefficients: " + str(lrModel.coefficients))
print("Intercept: " + str(lrModel.intercept))
```
